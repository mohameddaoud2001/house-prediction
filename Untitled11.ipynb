```python
import matplotlib.pyplot as plt
import numpy as np

# Define expense categories and their corresponding values
categories = ['Housing', 'Transportation', 'Food', 'Entertainment', 'Utilities']
expenses = [1500, 600, 800, 300, 400]

# Calculate the total expense
total_expense = sum(expenses)

# Calculate the angle for each category
angles = np.linspace(0, 2 * np.pi, len(categories), endpoint=False).tolist()
angles += angles[:1]

# Create a figure and axis
fig, ax = plt.subplots(subplot_kw={'polar': True})

# Plot the expenses as a radial plot
ax.fill(angles, expenses + [expenses[0]], 'b', alpha=0.1)

# Set labels for each category
ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories)

# Set the title and display the plot
plt.title('Household Budget Distribution')
plt.show()
```
